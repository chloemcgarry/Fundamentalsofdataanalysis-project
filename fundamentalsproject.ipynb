#Tips Data Set

Description:
A waiter's tips can vary dependant on many factors, such as part size and gender. For this project I will analyse the data set for relationships and use the Python packages seaborn and jupyter to do so.

```python
#importing the packages needed to run the data set
import numpy as np #importing numPy package
import pandas as pd #importing pandas package
import matplotlib.pyplot as plt #importing matplotlib package
import seaborn as sns #importing seaborn package
sns.set(style="darkgrid")
```

```python
#importing the tips data set from seaborn
tips = sns.load_dataset("tips")
```

```python
#The array that describes the data set
tips.shape
```

```python
tips.columns #data that is in the data set
```

```python
#Describe the data set - the number of data points, mean, standard deviation, min, max values.
tips.describe()
```

```python
#The first 5 data points in the data set
tips.head()
```

```python
#The last 5 data points in the data set
tips.tail()
```

```python
#analyzing the males tips
male = tips[tips.sex=='Male']
male.describe()
```

```python
#finding the number of males in the data set
malecount = int(male.describe().tip.iloc[0])
malecount
```

```python
#analyzing female tips
female = tips[tips.sex=='Female']
female.describe()
```

```python
#finding the number of females in the data set
femalecount = int(female.describe().tip.iloc[0])
femalecount
```
