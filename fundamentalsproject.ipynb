#Analysis

The 'Tips' data set analyses the tip given to a waiter by 244 people over a 4 day period in a restaurant. The waiter analyzes records different factors that may affect the amount of money given as a tip. These factors are the total bill, tip, smoker/non-smoker, sex, day the person vistited the restaurant and the size of the table (number of people at the table or covers). 

157 of the 244 entries were male which accounts for aprroximately 64.3% of the entries, while the remaining 87 were female making up an approximate 35.7% of the data. The average bill paid by a male was $ 20.744076 and $18.056897 for females. The males tipped on average of $ 3.08 and females was $ 2.83. The maximum and minimum tip given by males were $ 10.0 and $1.0 respectively. The maximum and minimum tip given by females were $ 6.5 and $1.0 respectively.

The data for the 'Tips' data set was collected across four different days - Thursday, Friday, Saturday and Sunday. It is unclear why these days were chosen. There could be many factors that contribute to why these days were chosen such as the restaurant does not open Monday - Wednesday or the waiter only works Thursday to Sunday. The restaurants busiest day was on Saturday - there were 87 data points recorded for Saturday, whereas Friday was the quietest day with only 19 data points recorded.
